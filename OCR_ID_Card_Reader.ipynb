```python
from PIL import Image
import pytesseract
import re

# Görseli yükleyin (dosya adını uploaded dict'ten alıyoruz)
image_path = list(uploaded.keys())[0]
image = Image.open('/content/WhatsApp Image 2024-06-26 at 15.54.29.jpeg')

# Görselden metni çıkar
text = pytesseract.image_to_string(image, lang='tur')

# Çıkarılan metni yazdır
print("Çıkarılan Metin:")
print(text)

# Kimlik bilgilerini düzenli ifadeler kullanarak çıkar
kimlik_bilgileri = {}

# TC Kimlik No (örnek düzenli ifade, TC Kimlik No 11 hanelidir)
tc_kimlik_no = re.search(r'\b\d{11}\b', text)
if tc_kimlik_no:
    kimlik_bilgileri['TC Kimlik No'] = tc_kimlik_no.group()

# Ad Soyad (örnek düzenli ifade, ad ve soyadın baş harfleri büyük olabilir)
ad_soyad = re.search(r'Ad Soyad: (.+)', text)
if ad_soyad:
    kimlik_bilgileri['Ad Soyad'] = ad_soyad.group(1)

# Doğum Tarihi (örnek düzenli ifade, tarih formatı gg.aa.yyyy)
dogum_tarihi = re.search(r'Doğum Tarihi: (\d{2}\.\d{2}\.\d{4})', text)
if dogum_tarihi:
    kimlik_bilgileri['Doğum Tarihi'] = dogum_tarihi.group(1)

# Diğer kimlik bilgileri için benzer düzenli ifadeler ekleyebilirsiniz
# ...

# Kimlik bilgilerini yazdır
print("Kimlik Bilgileri:")
for key, value in kimlik_bilgileri.items():
    print(f"{key}: {value}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Görseli cv2 ile yükleyin
image_cv2 = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Kimlik bilgilerini görselin üzerine yazın
start_y = 30
for i, (key, value) in enumerate(kimlik_bilgileri.items()):
    text = f"{key}: {value}"
    (text_width, text_height), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 1, 2)

    # Metin kutusunun sol üst ve sağ alt köşelerini hesaplayın
    top_left = (10, start_y - text_height)
    bottom_right = (10 + text_width, start_y + 10)

    # Kare çizin
    cv2.rectangle(image_cv2, top_left, bottom_right, (0, 255, 0), 2)

    # Metni yazın
    cv2.putText(image_cv2, text, (10, start_y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Bir sonraki metin satırı için y konumunu ayarlayın
    start_y += text_height + 40

# Görseli görüntüleyin
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
